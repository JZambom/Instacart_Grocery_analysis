```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = r'C:\Users\Jose Zambom\OneDrive - Exel Industries\Data Analysis\Data Analytics Immersion\Python Fundamentals for Data Analyst\csv file\Prepared Data\combined_data.pkl'
df = pd.read_pickle(data_path)

# Define a function to save and show plots
def save_and_show_plot(filename):
    plt.tight_layout()
    plt.savefig(f'C:/Users/Jose Zambom/OneDrive - Exel Industries/Data Analysis/Data Analytics Immersion/Python Fundamentals for Data Analyst/csv file/Visualization/{filename}')
    plt.show()
```

```python
# 1. Busiest days of the week
plt.figure(figsize=(10,6))
df['orders_day_of_week'].value_counts().sort_index().plot(kind='bar', color='blue')
plt.title('Busiest Days of the Week (Number of Orders)')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Orders')
save_and_show_plot('busiest_days_of_week.png')
```

```python
# 2. Times when people spend the most money
plt.figure(figsize=(10,6))
df.groupby('order_hour_of_day')['prices'].mean().plot(kind='line', marker='o', color='blue')
plt.title('Average Spending by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Spending ($)')
save_and_show_plot('average_spending_by_hour.png')
```

```python
# Define the path for the departments_wrangled file with .xlsx extension
departments_path = r'C:\Users\Jose Zambom\OneDrive - Exel Industries\Data Analysis\Data Analytics Immersion\Python Fundamentals for Data Analyst\csv file\Prepared Data\departments_wrangled.csv'

# Load the departments data
df_departments = pd.read_csv(departments_path)

# Check the first few rows to confirm the data
print(df_departments.head())
```

```python
# Check columns in the main dataframe
print("Columns in the main dataframe (df):", df.columns)

# Check columns in the departments dataframe
print("Columns in the departments dataframe (df_departments):", df_departments.columns)
```

```python
# Rename 'Unnamed: 0' to 'department_id' in df_departments
df_departments = df_departments.rename(columns={'Unnamed: 0': 'department_id'})

# Merge the main data with department names on 'department_id'
df_combined = pd.merge(df, df_departments, on='department_id', how='left')

# Group by department to count the frequency of orders
department_order_counts = df_combined['department'].value_counts()

# Plotting the results
plt.figure(figsize=(12, 6))
sns.barplot(x=department_order_counts.index, y=department_order_counts.values, palette="viridis")
plt.title('Frequency of Product Orders by Department')
plt.xlabel('Department')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()

# Save and show the plot
plt.savefig(r'C:\Users\Jose Zambom\OneDrive - Exel Industries\Data Analysis\Data Analytics Immersion\Python Fundamentals for Data Analyst\csv file\Visualization\product_orders_by_department.png')
plt.show()
```

```python
# Calculate the distribution of users by loyalty flag
loyalty_distribution = df['loyalty_flag'].value_counts()

# Plotting the distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=loyalty_distribution.index, y=loyalty_distribution.values, palette="pastel")

# Add titles and labels
plt.title('Distribution of Users by Loyalty Status')
plt.xlabel('Loyalty Status')
plt.ylabel('Number of Users')

# Save and show the plot
plt.savefig(r'C:\Users\Jose Zambom\OneDrive - Exel Industries\Data Analysis\Data Analytics Immersion\Python Fundamentals for Data Analyst\csv file\Visualization\loyalty_distribution.png')
plt.show()
```

```python
import os

# Define the path to save the images
visualization_folder = r'C:\Users\Jose Zambom\OneDrive - Exel Industries\Data Analysis\Data Analytics Immersion\Python Fundamentals for Data Analyst\csv file\Visualization'

# Step 1: Group by 'loyalty_flag' to calculate metrics
loyalty_order_metrics = df.groupby('loyalty_flag').agg(
    avg_order_size=('add_to_cart_order', 'mean'),
    avg_num_orders=('order_number', 'mean'),
    avg_spending_per_order=('prices', 'mean')
).reset_index()

# Step 2: Plot the metrics for visual analysis

import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots for each metric
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Average Order Size
sns.barplot(x='loyalty_flag', y='avg_order_size', data=loyalty_order_metrics, ax=axes[0], palette="Blues")
axes[0].set_title('Average Order Size by Loyalty Status')
axes[0].set_xlabel('Loyalty Status')
axes[0].set_ylabel('Average Order Size')

# Save the plot for average order size
avg_order_size_path = os.path.join(visualization_folder, 'avg_order_size_by_loyalty_status.png')
fig.savefig(avg_order_size_path)

# Average Number of Orders
sns.barplot(x='loyalty_flag', y='avg_num_orders', data=loyalty_order_metrics, ax=axes[1], palette="Oranges")
axes[1].set_title('Average Number of Orders by Loyalty Status')
axes[1].set_xlabel('Loyalty Status')
axes[1].set_ylabel('Average Number of Orders')

# Save the plot for average number of orders
avg_num_orders_path = os.path.join(visualization_folder, 'avg_num_orders_by_loyalty_status.png')
fig.savefig(avg_num_orders_path)

# Average Spending per Order
sns.barplot(x='loyalty_flag', y='avg_spending_per_order', data=loyalty_order_metrics, ax=axes[2], palette="Greens")
axes[2].set_title('Average Spending per Order by Loyalty Status')
axes[2].set_xlabel('Loyalty Status')
axes[2].set_ylabel('Average Spending ($)')

# Save the plot for average spending per order
avg_spending_per_order_path = os.path.join(visualization_folder, 'avg_spending_per_order_by_loyalty_status.png')
fig.savefig(avg_spending_per_order_path)

# Display the plots
plt.tight_layout()
plt.show()

# Save the final plot as a combined image in the visualization folder
combined_plot_path = os.path.join(visualization_folder, 'loyalty_order_metrics_combined.png')
fig.savefig(combined_plot_path)
```

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Define the save path for the charts
save_path = r'C:\Users\Jose Zambom\OneDrive - Exel Industries\Data Analysis\Data Analytics Immersion\Python Fundamentals for Data Analyst\csv file\Visualization'

# Function to save the figures
def save_figure(fig, filename):
    file_path = os.path.join(save_path, filename)
    fig.savefig(file_path)

# Define the age group categories
def classify_age(age):
    if age < 35:
        return 'Young Adult'
    elif 35 <= age <= 54:
        return 'Adult'
    elif 55 <= age <= 64:
        return 'Older Adult'
    else:
        return 'Senior'

# Apply the function to classify age into groups
df['age_group'] = df['age'].apply(classify_age)

# Group the data by age_group and family_status to calculate average order size, number of orders, and average expenditure
age_group_order_stats = df.groupby(['age_group', 'family_status']).agg({
    'order_number': 'mean',  # Average number of orders
    'add_to_cart_order': 'mean',  # Average order size
    'prices': 'mean'  # Average expenditure per order
}).reset_index()

# Create a bar chart to analyze the connection between age group and family status on the average number of orders
plt.figure(figsize=(10, 6))
sns.barplot(data=age_group_order_stats, x='age_group', y='order_number', hue='family_status')
plt.title('Average Number of Orders by Age Group and Family Status')
plt.xlabel('Age Group')
plt.ylabel('Average Number of Orders')
plt.legend(title='Family Status')
plt.tight_layout()

# Save the figure
save_figure(plt, 'average_number_of_orders_age_group_family_status.png')
plt.show()

# Create a bar chart for average order size by age group and family status
plt.figure(figsize=(10, 6))
sns.barplot(data=age_group_order_stats, x='age_group', y='add_to_cart_order', hue='family_status')
plt.title('Average Order Size by Age Group and Family Status')
plt.xlabel('Age Group')
plt.ylabel('Average Order Size')
plt.legend(title='Family Status')
plt.tight_layout()

# Save the figure
save_figure(plt, 'average_order_size_age_group_family_status.png')
plt.show()

# Create a bar chart for average expenditure by age group and family status
plt.figure(figsize=(10, 6))
sns.barplot(data=age_group_order_stats, x='age_group', y='prices', hue='family_status')
plt.title('Average Expenditure by Age Group and Family Status')
plt.xlabel('Age Group')
plt.ylabel('Average Expenditure ($)')
plt.legend(title='Family Status')
plt.tight_layout()

# Save the figure
save_figure(plt, 'average_expenditure_age_group_family_status.png')
plt.show()

```
